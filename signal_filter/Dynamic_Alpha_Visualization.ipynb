```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# make numbers from 0-2
x = np.linspace(0, 1, 100)

parameters = np.linspace(0.1, 1, 10)

# make y = x/parameter
y = np.zeros((len(parameters), len(x)))
for i, p in enumerate(parameters):
    y[i] = x / p

# limit each y to 1
y[y > 1] = 1

# plot the different y values
for i, p in enumerate(parameters):
    plt.plot(x, y[i], label='p = {:1.1f}'.format(p))

plt.xlabel('Difference in climb rate')
plt.ylabel('Dynamic alpha')

# start x and y from zero to 1 and make a grid
plt.axis([0, 1, 0, 1])
plt.grid(True)

# make a dotted vertical line at x=0.1
plt.axvline(x=0.1, color='k', linestyle=':', lw=1)

plt.title('Dynamic Alpha vs Difference in Climb Rate')

plt.legend()
plt.show()
```
